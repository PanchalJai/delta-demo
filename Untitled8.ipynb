```python
from types import CellType
import random
# Tic Tac Toe Game in Python

# Function to print the board

def print_board(board):
  for row in board:
    print("|".join(row))
    print("-"*9)


# function to chack for a winner
def check_winner(board, player):
  # check rows
  for row in board:
    if all(CellType == player for CellType in row):
      return True
  # check columns
  for col in range(3):
    if all(board[row][col] == player for row in range(3)):
      return True

  # check diagonals
  if all(board[i][i] == player for i in range(3)):
    return True
  if all(board[i][2-i] == player for i in range(3)):
    return True
  return False


# function to check if board is full
def is_board_full(board):
  return all(CellType != " " for row in board for CellType in row)

# function for "o" tern
def computer_turn(board):
  while True:
    row = random.randint(0, 2)
    col = random.randint(0, 2)
    if board[row][col] == " ":
      board[row][col] = "O"
      print("Computer's turn:")
      print_board(board)
      break
    else:
      computer_turn(board)
      break

# main game loop
def play_game():
  board = [[" " for _ in range(3)] for _ in range(3)]
  players = ["X", "O"]
  turns = 0

  print("welcome to tic tac toe")
  print_board(board)


  while True:

    player = players[0]
    print(f"\nplayer {player}'s turn")

    # Get velid input
    while True:
      try:
        row = int(input("Enter row (0, 1, 2): "))
        col = int(input("Enter column (0, 1, 2): "))
        if  board[row][col] == " ":
          board[row][col] = player
          break

        else:
          print("That cell is already taken. Try again.")
      except ValueError:
        print("Invalid input. Enter a number between 0 and 2.")

    print_board(board);

    computer_turn(board)      # For computer tern1
    0



    # check winner
    if check_winner(board, player):
      print(f"Player {player} wins!")
      break

    turns += 1


# Run the Game
play_game()

```

```python
import math

# Initialize the board
board = [' ' for _ in range(9)]

# Function to print the board
def print_board(board):
  print()
  print(f"{board[0]} | {board[1]} | {board[2]}")
  print("---+---+---")
  print(f"{board[3]} | {board[4]} | {board[5]}")
  print("---+---+---")
  print(f"{board[6]} | {board[7]} | {board[8]}")
  print()

# check for winner
def check_winner(board,player):
  winning_combinations = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                          (0,3,6), (1,4,7),(2,5,8),
                          (0,4,8), (2,4,6)]
  return any(board[a]==board[b]==board[c] == player for a,b,c in winning_combinations)

#  check for draw
def is_draw(board):
  return ' ' not in board


# eveluate board for minmax
def evaluate(board):
  if check_winner(board, 'O'):   # computer wins
    return 1
  elif check_winner(board, 'X'):  # Human Wins
    return -1
  else:
    return 0                      # Drow or ongoing


# Minmax Function
def minimax(board, depth, is_maximizing ):
  score = evaluate(board)

  # Terminal State
  if score ==1 or score == -1:
    return score
  if is_draw(board):
    return 0

  if is_maximizing:
    best = -math.inf
    for i in range(9):
      if board[i]== ' ':
        board[i] = 'O'
        best =max(best, minimax(board, depth+1, False))
        board[i] = ' '
    return best
  else:
    best = math.inf
    for i in range(9):
      if board[i]== ' ':
        board[i] = 'X'
        best = min(best, minimax(board, depth+1, True))
        board[i] = ' '
    return best

# Find best move for computer
def best_move(board):
  best_val = -math.inf
  move = -1
  for i in range(9):
    if board[i] == ' ':
      board[i] = 'O'
      move_val = minimax(board, 0, False)
      board[i] = ' '
      if move_val > best_val:
        best_val = move_val
        move = i
  return move

# Main game function
def play_game():
  print("Welcome to Tic Tac Toe!")
  print("You are 'X' and the computer is 'O'.")
  print_board(board)

  while True:
    # Human Move
    try:
      move = int(input("Enter your move (0-8): "))
    except ValueError:
      print("Invalid input. Enter a number between 0 and 8.")
      continue
    if move < 0 or move > 8 or board[move]!= ' ':
      print("Invalid input. Enter a number between 0 and 8.")
      continue
    board[move] = 'X'
    print_board(board)

    if check_winner(board, 'X'):
      print("You win!")
      break

    if is_draw(board):
      print("It is drow!")
      break

    # Computer Move
    print("Computer's turn...")
    move = best_move(board)
    board[move] = 'O'
    print_board(board)

    if check_winner(board, 'O'):
      print("Computer wins!")
      break

    if is_draw(board):
      print("It is drow!")
      break

# Run the game
play_game()


```
